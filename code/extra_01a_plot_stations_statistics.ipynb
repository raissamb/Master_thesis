## Create a total occupation table

```python
# Create a new dataframe with info from each region calculated previously
frames = [n_df, ne_df, co_df, se_df, s_df]
dff = pd.concat(frames)
ndf = dff.reset_index(drop=True)
```

```python
# Add columns
total_sts = ndf["Stations"].sum()
total_ocp = ndf["Occupations"].sum()
```

```python
# append to dataframe
new_row = {"States": "Total", "Stations": total_sts, "Occupations": total_ocp}
ndf = ndf.append(new_row, ignore_index=True)

# Visualize
ndf
```

```python
# Save file
ndf.to_csv(output_folder / table_total, index=False)
#Total_ocp = df_final.sum(level=1)
```

### Plot the total tables

```python
# Define font size
bar_label_size = 14
title_size = 26
xlabel_size = 18
ylabel_size = 18
xticks_size = 14
yticks_size = 14
```

```python
# Create a dataframe with crescent order of number of stations
crescent_ndf_stations = ndf.sort_values(by="Stations", ascending=True)

# Plot total stations table
figure(num=None, figsize=(20,18), dpi=80, facecolor='w', edgecolor='r')
plot_total_station = sns.barplot(x="Stations", y="States", data=crescent_ndf_stations, color="blue", ci=None) # ci none to remove error bar

# bar_label() function to annotate over the bars
plt.bar_label(plot_total_station.containers[0], size = bar_label_size)

# Details
plt.title("Total number of stations by States", size = title_size)
plt.xlabel("Number of stations", size = xlabel_size)
plt.ylabel("States", size = ylabel_size)

# Axis size
plt.xticks(fontsize = xticks_size)
plt.yticks(fontsize = yticks_size)

# Save fig
plt.savefig(output_folder / total_stations)
```

```python
# Create a dataframe with crescent order of number of occupations
crescent_ndf_ocp = ndf.sort_values(by="Occupations", ascending=True)

# Plot total reocp table
figure(num=None, figsize=(20,18), dpi=80, facecolor='w', edgecolor='r')
plot_total_ocp = sns.barplot(x="Occupations", y="States", data=crescent_ndf_ocp, color="blue", ci = None)

# bar_label() function to annotate over the bars
plt.bar_label(plot_total_ocp.containers[0], size = bar_label_size)

# Details
plt.title("Total number of occupations by States", size = title_size)
plt.xlabel("Number of occupations", size = xlabel_size)
plt.ylabel("States", size = ylabel_size)

# Axis size
plt.xticks(fontsize = xticks_size)
plt.yticks(fontsize = yticks_size)

# Save fig
plt.savefig(output_folder / total_occupations)
```
