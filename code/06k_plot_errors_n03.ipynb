# Plot the errors for each occupation group: n 03

## Description

This program does the following using the Repeat station and IGRF error database:
- It reads the file with the dataframe from the n occupation stations group
- It plots the RMSE, Mean NRMSE and Std NRMSE values for each group

```python
# Import modules
import mestrado_module as mm
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pathlib import Path
```

```python
## Define the minimum number of occupations a station must have (n), it is a cut out value
n1 = 3
n2 = 5

# File with the chosen n repeat stations
errors_df_folder: Path = Path(mm.path_pipeline_04_igrf_calc)
errors_df_file: Path = Path(mm.output_4d_code_error_database)

# Save figures files
output_folder: Path = Path(mm.path_pipeline_06e_temporal_series_n03)

# Save figures files for rmse

## Group 1
g1_fig_rmse_f: Path = Path("00_g1_rmse_f_plot_n_03.png")
g1_fig_rmse_h: Path = Path("00_g1_rmse_h_plot_n_03.png")
g1_fig_rmse_x: Path = Path("00_g1_rmse_x_plot_n_03.png")
g1_fig_rmse_y: Path = Path("00_g1_rmse_y_plot_n_03.png")
g1_fig_rmse_z: Path = Path("00_g1_rmse_z_plot_n_03.png")
g1_fig_rmse_d: Path = Path("00_g1_rmse_d_plot_n_03.png")
g1_fig_rmse_i: Path = Path("00_g1_rmse_i_plot_n_03.png")

## Group 2
g2_fig_rmse_f: Path = Path("00_g2_rmse_f_plot_n_03.png")
g2_fig_rmse_h: Path = Path("00_g2_rmse_h_plot_n_03.png")
g2_fig_rmse_x: Path = Path("00_g2_rmse_x_plot_n_03.png")
g2_fig_rmse_y: Path = Path("00_g2_rmse_y_plot_n_03.png")
g2_fig_rmse_z: Path = Path("00_g2_rmse_z_plot_n_03.png")
g2_fig_rmse_d: Path = Path("00_g2_rmse_d_plot_n_03.png")
g2_fig_rmse_i: Path = Path("00_g2_rmse_i_plot_n_03.png")

# Figure style
sns.set_style("darkgrid")
```

## Read the data

```python
df = pd.read_csv(errors_df_folder / errors_df_file)
df.info()
```

## Plot the temporal series for each station in this n group

```python
# Define variables
component_f = "Total Field"
component_h = "Horizontal Field"
component_x = "X Field"
component_y = "Y Field"
component_z = "Z Field"
component_d = "Declination Field"
component_i = "Inclination Field"

# Original
orig_symbol = "o"
orig_color = "blue"
orig_label = "Station original value"
orig_linestyle = '-'

# Calculated
calc_symbol = "+"
calc_color = "red"
calc_label = "Station calculated value"
calc_linestyle = "--"

# IGRF
igrf_symbol = "d"
igrf_color = "green"
igrf_label = "IGRF13"
igrf_linestyle = "-."

# General
axis_label_fsize = 14
title_fontsize = 16
tick_size = 14
legend_loc = "best"
dpi_quality = 300
value_bbox_inches = "tight"
fill_color = "purple"
```

```python
# Auxiliar df
aux_df = df.loc[df["N_occupations"].between(3,5)]
aux_df = aux_df.reset_index()
```

```python
# Define intervals
int0 = 0
int1 = 20
int2 = 21
int3 = 38
```

```python
# Group
g1_df = aux_df.loc[int0:int1]
g2_df = aux_df.loc[int2:int3]
#view
g1_df
```

## Plot the RMSE

### Group 1

```python
# F field info for plot
rmse_f = g1_df["RMSE_F_Original_values"]
station_name = g1_df["Code"]
# Figure
fig1, ax1 = plt.subplots()
ax1.scatter(station_name, rmse_f, marker = orig_symbol)
ax1.set_xlabel("Station name", fontsize=axis_label_fsize)
ax1.set_ylabel("RMSE Value", fontsize=axis_label_fsize)
ax1.set_title(f"{component_f}: Group with {n1} or {n2} or more occupations", fontsize=title_fontsize)
plt.legend([orig_label, calc_label])
plt.xticks(rotation=90)
plt.savefig(output_folder / g1_fig_rmse_f, dpi=dpi_quality, bbox_inches=value_bbox_inches)
plt.show()
```
